```python
def convert_data(input_file, output_file):
    with open(input_file, 'r') as f:
        lines = f.readlines()

    name = ''
    comments = ''
    dim = ''
    capacity = ''
    optimal = ''
    nodes = []
    demand_dict = {}

    nodes_start = False
    demand_start = False

    for line in lines:
        if line.startswith('NAME :'):
            name = line.split(': ')[1].strip()
        elif line.startswith('COMMENT :'):
            comments = line.split(': ')[1].split(', ')[0].strip()
            optimal = line.split('Optimal value: ')[1].strip().split()[0].strip(')')
        elif line.startswith('DIMENSION :'):
            dim = line.split(': ')[1].strip()
        elif line.startswith('CAPACITY :'):
            capacity = line.split(': ')[1].strip()
        elif line.startswith('NODE_COORD_SECTION'):
            nodes_start = True
        elif line.startswith('DEMAND_SECTION'):
            nodes_start = False
            demand_start = True
        elif line.startswith('DEPOT_SECTION'):
            demand_start = False
        elif nodes_start and line.strip():
            nodes.append(line.strip())
        elif demand_start and line.strip():
            node_demand = line.strip().split()
            demand_dict[node_demand[0]] = node_demand[1]

    with open(output_file, 'w') as f:
        f.write(f'NAME: {name}\n')
        f.write(f'COMMENTS: {comments} {optimal} A Set\n')
        f.write(f'DIM: {dim}\n')
        f.write(f'CAPACITY: {capacity}\n')
        f.write(f'OPTIMAL: {optimal}\n')
        f.write(f'NODES:\n')
        for node in nodes:
            node_data = node.split()
            node_id = node_data[0]
            if node_id in demand_dict:
                f.write(f'{node} {demand_dict[node_id]}\n')
            else:
                f.write(f'{node} 0\n')

convert_data('A-n32-k5.vrp', 'A-n32-k5.ocvrp')
```

```python
import os
import glob

# Используем glob для поиска всех файлов .vrp в директории
vrp_files = glob.glob('*.vrp')

# Применяем функцию convert_data к каждому файлу и удаляем исходные файлы .vrp
for vrp_file in vrp_files:
    ocvrp_file = vrp_file.replace('.vrp', '.ocvrp')
    convert_data(vrp_file, ocvrp_file)
    os.remove(vrp_file)
```
